```python
%matplotlib inline 

import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
PATH = "../data/raw/"

FILE_NETWORK_SOCIAL_EXAMPLE = "higgs-social_network_example.edgelist"
FILE_NETWORK_RETWEET_WEIGHT_EXAMPLE = "higgs-retweet_network_example.edgelist"
```

```python
data = pd.read_csv(PATH + FILE_NETWORK_RETWEET_WEIGHT_EXAMPLE, sep=' ', names = ['source', 'target','weight'])
data.head()
```

```python
G = nx.from_pandas_dataframe(data, 'source', 'target')
```

```python
#k_core(G, k=None, core_number=None)[source]
#Return the k-core of G.
#A k-core is a maximal subgraph that contains nodes of degree k or more.

k_core_graph = nx.k_core(G)
nx.draw(k_core_graph, with_labels = True, node_size=700)
plt.show()
```

```python
df2 = nx.to_pandas_dataframe(k_core_graph)
df2
```

```python
#k_core(G, k=None, core_number=None)[source]
#Return the k-core of G.
#A k-core is a maximal subgraph that contains nodes of degree k or more.

k_shell_graph = nx.k_shell(G)
nx.draw(k_shell_graph, with_labels = True, node_size=700)
plt.show()
```

```python
df3 = nx.to_pandas_dataframe(k_shell_graph)
df3
```
